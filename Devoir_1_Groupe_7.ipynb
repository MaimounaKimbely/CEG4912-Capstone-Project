# Algorithm for generate and test

## Ici on utilise 2 fonctions (nextPermutation() et knapSack()).

## nextPermutation() deplace l'article deja dans le sac en fin de liste et tente de trouver un autre item qui fit

## knapSack() ajoute l'item si son poids ajouté au poids total n'excède pas la capacité et crée une liste chargée d'afficher le meilleur résulat avec 1 si l'objet est présent dans le sac et 0 dans le cas contraire


```python
INT_MIN = -510

def nextPermutation(nums: list) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    if sorted(nums, reverse=True) == nums:
        return None
    n = len(nums)
    brk_point = -1
    for pos in range(n - 1, 0, -1):
        if nums[pos] > nums[pos - 1]:
            brk_point = pos
            break
    else:
        nums.sort()
        return

    replace_with = -1
    for j in range(brk_point, n):
        if nums[j] > nums[brk_point - 1]:
            replace_with = j
        else:
            break
    nums[replace_with], nums[brk_point - 1] = nums[brk_point - 1], nums[replace_with]
    nums[brk_point:] = sorted(nums[brk_point:])
    return nums

# Function to find all the possible solutions of the
# 0/1 knapSack problem
def knapSack(W, wt, val, n):
    # Mapping weights with Profits
    umap = dict()

    set_sol = set()
    included_weights = [0] * n  # Initialize a list to store included weights

    # Making Pairs and inserting to the map
    for i in range(n):
        umap[wt[i]] = val[i]

    result = INT_MIN
    remaining_weight = 0
    sum = 0

    # Loop to iterate over all the
    # possible permutations of array
    while True:
        sum = 0

        # Initially, the bag will be empty
        remaining_weight = W
        possible = []

        # Loop to fill up the bag
        # until there is no weight
        # such which is less than
        # the remaining weight of the
        # 0-1 knapsack
        for i in range(n):
            if wt[i] <= remaining_weight:
                remaining_weight -= wt[i]
                sum += umap[wt[i]]
                possible.append((wt[i], umap[wt[i]]))
                included_weights[i] = 1  # Mark the weight as included

        possible.sort()
        if sum > result:
            result = sum

        if tuple(possible) not in set_sol:
            for sol in possible:
                print(sol[0], ": ", sol[1], ", ", end='')
            print()
            set_sol.add(tuple(possible))

        if not nextPermutation(wt):
            break

    return result, included_weights

# Driver Code
if __name__ == '__main__':
    """
    val = data['Prices'][0]
    wt = data['Weights'][0]
    W =data['Capacity'][0]
    n = len(data['Prices'][0])


    val = [19, 16, 16,  4,  1]
    wt = [32, 49, 27, 37, 24]
    W = 64
    n = len(val)

    Best_price, included_weights = knapSack(W, wt, val, n)
    print("Best_price =", Best_price)
    print("Included Weights =", included_weights)
    print(type(Best_price))
    """

    #loop to go through the whole dataframe


    # Perform an operation to generate 1000 integers (e.g., generating numbers from 1 to 1000)
    generated_numbers = range(1, 10001)

    #9999 10000
    # Assign the generated numbers to the 'EmptyColumn'
    data['New_Best_Price'] = generated_numbers
    data['New_Best_Pick'] = [[0, 0, 0, 0, 0]] * len(data)


    """
    val = data['Prices'][1]
    wt = data['Weights'][1]
    W = data['Capacity'][1]
    n = len(data['Prices'][1])
    data['New_Best_Price'][1], data['New_Best_Pick'][1] = knapSack(W, wt, val, n)
    """


    for x in range(len(data['Prices'])):
        val = data['Prices'][x]
        wt = data['Weights'][x]
        W = data['Capacity'][x]
        n = len(data['Prices'][x])
        data['New_Best_Price'][x], data['New_Best_Pick'][x] = knapSack(W, wt, val, n)

```

## On peut expérimenter avec différentes valeurs pour facteur_refroidissement et iterations pour observer leur impact sur les résultats. Une valeur plus élevée de facteur_refroidissement refroidira plus lentement le système, ce qui pourrait permettre une exploration plus approfondie de l'espace de recherche, mais nécessitera également plus d'itérations.

## Un nombre plus élevé d'itérations pourrait conduire à de meilleures solutions, mais augmentera également le temps d'exécution. Il est important de trouver un équilibre entre ces paramètres en fonction de votre problème spécifique


```python
import random
import math

# Fonction de fitness : somme des valeurs des objets inclus dans le sac
def fitness(solution, valeurs):
    total_valeur = sum(valeurs[i] for i in range(len(solution)) if solution[i] == 1)
    return total_valeur

# Algorithme de recuit simulé
def recuit_simule(valeurs, poids, capacite, temperature_initiale, facteur_refroidissement, iterations):
    n = len(valeurs)
    solution_courante = [random.randint(0, 1) for _ in range(n)]
    fitness_courante = fitness(solution_courante, valeurs)

    meilleure_solution = solution_courante.copy()
    meilleure_fitness = fitness_courante

    temperature = temperature_initiale

    for _ in range(iterations):
        # Générer une nouvelle solution en effectuant une modification aléatoire
        index = random.randint(0, n - 1)
        nouvelle_solution = solution_courante.copy()
        nouvelle_solution[index] = 1 - nouvelle_solution[index]

        # Calculer la fitness de la nouvelle solution
        nouvelle_fitness = fitness(nouvelle_solution, valeurs)

        # Calculer la différence de fitness
        delta_fitness = nouvelle_fitness - fitness_courante

        # Accepter la nouvelle solution avec une certaine probabilité
        if delta_fitness > 0 or random.random() < math.exp(delta_fitness / temperature):
            solution_courante = nouvelle_solution
            fitness_courante = nouvelle_fitness

        # Mettre à jour la meilleure solution si nécessaire
        if fitness_courante > meilleure_fitness and sum(poids[i] for i in range(n) if solution_courante[i] == 1) <= capacite:
            meilleure_solution = solution_courante.copy()
            meilleure_fitness = fitness_courante

        # Refroidir la température
        temperature *= facteur_refroidissement

    return meilleure_solution, meilleure_fitness

# Exemple d'utilisation

for x in range(len(data)):
    valeurs = data['Prices'][x]
    poids = data['Weights'][x]
    capacite = data['Capacity'][x]
    temperature_initiale = 1000
    facteur_refroidissement = 0.95
    iterations = 1000

    data['Recuit_Best_Pick'][x], data['Recuit_fitness'][x] = recuit_simule(valeurs, poids, capacite, temperature_initiale, facteur_refroidissement, iterations)

    #print("Meilleure solution :", meilleure_solution)
    #print("Meilleure fitness :", meilleure_fitness)
```

## On peut expérimenter avec différentes valeurs pour facteur_refroidissement et iterations pour observer leur impact sur les résultats.

## Une valeur plus élevée de facteur_refroidissement refroidira plus lentement le système, ce qui pourrait permettre une exploration plus approfondie de l'espace de recherche, mais nécessitera également plus d'itérations.

## Un nombre plus élevé d'itérations pourrait conduire à de meilleures solutions, mais augmentera également le temps d'exécution. Il est important de trouver un équilibre entre ces paramètres en fonction de votre problème spécifique

```python
import random

# Function to initialize a population of individuals
def initialize_population(pop_size, n):
    return [[random.randint(0, 1) for _ in range(n)] for _ in range(pop_size)]

# Function to calculate the fitness of an individual (total value)
def fitness(individual, values):
    total_value = sum(values[i] for i in range(len(individual)) if individual[i] == 1)
    return total_value

# Function to select parents for reproduction (tournament selection)
def select_parents(population, values, capacity):
    # Perform tournament selection
    tournament_size = 5
    selected_parents = []
    for _ in range(len(population)):
        tournament = random.sample(population, tournament_size)
        tournament_fitness = [fitness(ind, values) for ind in tournament]
        max_fitness_index = tournament_fitness.index(max(tournament_fitness))
        selected_parents.append(tournament[max_fitness_index])
    return selected_parents

# Function to perform single-point crossover
def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

# Function to perform mutation
def mutate(individual, mutation_rate):
    mutated_individual = individual.copy()
    for i in range(len(individual)):
        if random.random() < mutation_rate:
            mutated_individual[i] = 1 - mutated_individual[i]
    return mutated_individual

# Genetic algorithm for Knapsack problem
def genetic_knapsack(values, weights, capacity, pop_size, generations):
    n = len(values)
    population = initialize_population(pop_size, n)

    for _ in range(generations):
        parents = select_parents(population, values, capacity)
        new_population = []

        while len(new_population) < pop_size:
            parent1, parent2 = random.choice(parents), random.choice(parents)
            child1, child2 = crossover(parent1, parent2)
            child1 = mutate(child1, mutation_rate=0.1)
            child2 = mutate(child2, mutation_rate=0.1)
            new_population.extend([child1, child2])

        population = new_population

    best_individual = max(population, key=lambda ind: fitness(ind, values))
    best_fitness = fitness(best_individual, values)
    return best_individual, best_fitness

# Example usage



pop_size = 100
generations = 100

#for x in range(len(data['Prices'])):
values = data['Prices'][0]
weights = data['Weights'][0]
capacity = data['Capacity'][0]

data['Genetique_Best_Pick'][0], data['Genetique_fitness'][0] = genetic_knapsack(values, weights, capacity, pop_size, generations)

#print("Best Solution:", best_solution)
#print("Best Fitness:", best_fitness)
```
